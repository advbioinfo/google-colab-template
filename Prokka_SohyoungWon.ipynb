## 04. Visualization

Install CGView for visualization

```python
!conda install -c bioconda cgview
```

Parse GFF file to CGView input format.

```python
gff = open('prokka_outputs/test_plasmid.gff', 'r')
outfile = open('test_plasmid.tab', 'w')

gff.readline()
length = gff.readline().strip().split()[-1]

outfile.write('#test_plasmid\n')  # First line: '#' character followed by the name of the sequence.
outfile.write('%'+length+'\n')  # Second line: '%' character followed by the length of the sequence
outfile.write('!strand\tslot\tstart\tstop\ttype\n') # Header starting with '!' character. strand, slot, start, stop, type 

for line in gff:
  if line.startswith('##FASTA'):
    break # The input fasta is at the end of the gff file. Remove this part.
  temp = line.split('\t')
  # Get the relevant information from the gff line. 
  # Give a dummy number for slot. 
  # CGView does not get 'CDS' for type so change to 'gene'.
  outline = '\t'.join([temp[6], '1', temp[3], temp[4], 'gene'])+'\n'
  outfile.write(outline)


gff.close()
outfile.close()
```

Run GCView

```python
!cgview -i test_plasmid.tab -o test_plasmid_annotated.png
```
